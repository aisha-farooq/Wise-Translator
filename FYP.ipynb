```python
import numpy as np
from torchtext.data.metrics import bleu_score
from nltk.translate.bleu_score import SmoothingFunction, sentence_bleu
from deep_translator import GoogleTranslator
```

```python

translator = GoogleTranslator(source='en', target='ur')

f=open("data/EngTest.txt", encoding="utf8")
data = f.read().split('\n')
f.close()
f=open("data/UrdTest.txt", encoding="utf8")
data2 = f.read().split('\n')
f.close()
```

```python
original =[]
translations =[]
scores=[]
for src, trg in zip(data,data2):
  phrase = translator.translate(src)
  original.append(trg)
  translations.append(phrase)
  cc = SmoothingFunction()
  try:
      score = sentence_bleu(phrase, trg, smoothing_function=cc.method4)
  except:
      score = bleu_score([phrase], [trg])

  # print(src)
  # # print('> ' + phrase + '\n')
  # # print(score)
        
  scores.append(score)
print('GOOGLE BLEU Score on Test Set: ', np.array(scores).mean()*100)
```
